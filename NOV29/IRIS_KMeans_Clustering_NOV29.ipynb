```python
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
iris_data = sns.load_dataset('iris')
iris_data.head(20)
```

```python
del iris_data['species']
```

```python
iris_data
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
scaled_iris = scaler.fit_transform(iris_data)
scaled_iris
```

```python
from matplotlib import pyplot as plt
from sklearn.cluster import KMeans
```

```python
kmeans_cluster = KMeans(n_clusters=3)
kmeans_cluster.fit(scaled_iris)
pred = kmeans_cluster.predict(scaled_iris)
```

```python
import pandas as pd
iris_data['Target'] = pd.DataFrame(pred,columns=['Target'])
iris_data
```

```python
iris_data[iris_data['Target']==2]
```

```python
wcss = []

for i in range(1,11):
    kmeans_model = KMeans(n_clusters=i)
    kmeans_model.fit(scaled_iris)
    wcss.append(kmeans_model.inertia_)
```

```python
wcss
```

```python
plt.plot(range(1,11),wcss)
```

```python
wcss = []

for i in range(1,10):
    kmeans = KMeans(n_clusters=i)
    kmeans.fit(scaled_iris)
    wcss.append(kmeans.inertia_) #wcss
   
```

```python
wcss = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i)#,random_state=0)
    kmeans.fit(scaled_iris)
    wcss.append(kmeans.inertia_) 
    print(i,wcss)
    #Inertia is the sum of squared error for each cluster. 
    #Therefore the smaller the inertia the denser the cluster(closer together all the points are) 

plt.plot(range(1, 11), wcss)
plt.title('Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```
